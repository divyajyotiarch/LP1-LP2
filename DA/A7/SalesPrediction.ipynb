```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np

df_train = pd.read_csv('/home/divyajyoti/Sales/train.csv')
df_train.head()
```

```python
df_train.replace({"":np.nan}, inplace=True)
df_train['Item_Weight']=df_train['Item_Weight'].replace(np.nan,0)
df_train.groupby('Outlet_Type')['Outlet_Size'].apply(lambda x: x.mode().iat[0])
```

```python
df_train['Outlet_Size']=df_train['Outlet_Size'].replace(np.nan,"Small")
df_train
```

```python
df_train.isna().sum()
```

```python
df_train["Item_Type"].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder

# enc = OneHotEncoder(handle_unknown='ignore',categories='auto')
# feature_array=enc.fit_transform(df_train[['Item_Type','Outlet_Size','Outlet_Type']]).toarray()
# feature_l = enc.categories_
# feature_l = np.array(feature_l).ravel()
# feature_l.shape
# features = pd.DataFrame([feature_l],columns=feature_l)
# # pd.get_dummies(df_train)
# features.head()
df_train = pd.concat([df_train,pd.get_dummies(df_train['Item_Type'], prefix='Item_Type')],axis=1).drop(['Item_Type'],axis=1)
df_train = pd.concat([df_train,pd.get_dummies(df_train['Outlet_Size'], prefix='Outlet_Size')],axis=1).drop(['Outlet_Size'],axis=1)
df_train = pd.concat([df_train,pd.get_dummies(df_train['Outlet_Type'], prefix='Outlet_Type')],axis=1).drop(['Outlet_Type'],axis=1)
df_train = pd.concat([df_train,pd.get_dummies(df_train['Item_Fat_Content'], prefix='Item_Fat_Content')],axis=1).drop(['Item_Fat_Content'],axis=1)
df_train = pd.concat([df_train,pd.get_dummies(df_train['Outlet_Location_Type'], prefix='Outlet_Location_Type')],axis=1).drop(['Outlet_Location_Type'],axis=1)
```

```python
df_train.head()
```

```python
y = df_train.Item_Outlet_Sales
x = df_train.drop('Item_Outlet_Sales',axis=1)
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.75,test_size=0.25)
x_train.head()
```
